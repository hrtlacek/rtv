# Real Time Visual Content Creation - Overview and Comparison
This notebook tries to visualize a dataset containing evaluation of different programmable environments for real time video content creation. Please also refer to https://github.com/hrtlacek/rtv

Some of this data is objective (does a specifix environment run natively on linux or not?) some of it is rather subjective and wishy-washy(is the documentation of language x now 0.8 or 0.7257?). So take it with a grain of salt. Discussion and collaboration via git very welcome.

````python
%pylab inline
import pandas as pd
dataset = pd.read_csv("https://github.com/hrtlacek/rtv/raw/master/envCompare.csv")
````

````python
dataset
````

````python
transposed = dataset.T.loc['learning Curve/Easyness':'DMX/Artnet Support']
transposed.columns = dataset['name']
transposed
````

````python
p = transposed.plot(figsize=[17,15],subplots=True, kind='bar')
plt.xticks=transposed.index
show()
````

````python
from mpl_toolkits.mplot3d import Axes3D
````

# 3d feature comparison
Below you can compare the environments in a 3d plot. You can choose which 3 features you want to compare via the numbers in third to fith line. You can just count the features from above (0 based!) and enter the number below. For example if you want to use if a software runs on macOS  for the x-axis you would set the line

```xfeat = list(transposed.index)[8]```

to


```xfeat = list(transposed.index)[5]```

After you have done that you need to execute the cell via ctrl-enter (or the little play button to the top left of the cell)


````python
fig = plt.figure(figsize=[10,10])

xfeat = list(transposed.index)[8] #change these numbers to plot oter features in 3d!
yfeat = list(transposed.index)[9]
zfeat = list(transposed.index)[11]

xs = dataset[xfeat]
ys = dataset[yfeat]
zs = dataset[zfeat]



ax = fig.gca(projection='3d')


ax.set_xlabel(xfeat)
ax.set_ylabel(yfeat)
ax.set_zlabel(zfeat)

for i in range(len(xs)):
  
  text = list(transposed.columns)[i]
  ax.text(xs[i], ys[i], zs[i], text)

for i in range(len(xs)):
    color = p[i].get_children()[0].get_facecolor()
    ax.plot([0,xs[i]], [ys[i], ys[i]], [zs[i], zs[i]], color=color)
    ax.plot([xs[i],xs[i]], [1, ys[i]], [zs[i], zs[i]], color=color)
    ax.plot([xs[i],xs[i]], [ys[i], ys[i]], [0, zs[i]], color=color)
  
  
ax.set_xlim(0,1)
ax.set_ylim(0,1)
ax.set_zlim(0,1)
ax.scatter(xs,ys,zs)
show()
````
